#  ADF-test on pair spread
**ADF** --> Augmented Dickey Fuller

* t = t-test statistic
* p = p-value (significant value = 0.05)
* n = number of lags

### Case studies

* [CS1 - Orkla / Danone](#Case-study-1---ORK-/-/BN)
* [CS2 - Elkem / Wacker](#Case-study-2---ELK-/-WCH)



```python
import yfinance as yf
import pandas as pd
from statsmodels.tsa.stattools import adfuller
```

### Case study 1 - ORK / BN

```python
s1 = yf.download('ORK.OL')['Close']['2020-01-01':]
s2 = yf.download('BN.PA')['Close']['2020-01-01':]

spread = s1 - s2
spread.dropna(inplace=True)
spread.plot()
```

```python
### adfuller for spread

t, p, n = adfuller(spread)[:3]

print(f't-statistic: {t:.3f}')
print(f'p-value: {p:.3f}')
print(f'number of lags: {n}')
```

```python
### Differentiating

spread_1d = spread.diff()[1:]
spread_1d.plot()
```

```python
### adfuller for differentiated spread

t, p, n = adfuller(spread_1d)[:3]

print(f't-statistic: {t:.3f}')
print(f'p-value: {p:.3f}')
print(f'number of lags: {n}')
```

It seems like the differentiated spread is stationary

```python
### Mean and STD
mean, std = spread_1d.mean(), spread_1d.std()
print(f'Mean of differentiated spread: {mean:.3f}')
print(f'std of differentiated spread: {std:.3f}')
```
